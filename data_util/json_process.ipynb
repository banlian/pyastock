```python

idlines = []

with open('json', 'r', encoding='utf-8') as fs:

    lines = fs.readlines()

    idlines = [l for l in lines if l.startswith('"id')]

print(idlines)
```

```python
import re

ids = []
for l in idlines:
    m = re.findall(':\"(.*)?\",', l)
    ids.append(m[0])

print(ids)
```

```python
codes = [c[:6] for c in ids]
print(codes)
```

```python
from stockbase.stock_db import db_id_to_name

dbids = [d[:6] for d in dbids]

ids = [d[:6] for d in ids]
ids2 = [d for d in ids if len(d) == 6]

print('db id', len(dbids))
print('json id', len(ids))

for id in dbids:
    if id not in ids2:
        print(id, db_id_to_name(int(id)))
```
